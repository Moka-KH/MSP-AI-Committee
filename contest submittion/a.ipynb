## I simplified

```python
threshold = 10
brand_counts = df["brand"].value_counts()
df["brand"] = df["brand"].apply(lambda x: x if brand_counts[x] > threshold else "Others")


df['brand'].value_counts()
```

```python
df = pd.get_dummies(df, columns=["brand"], dtype=int)
```
